### Standardizing the data, so that it would range from -1 to +1

```python
from sklearn.preprocessing import StandardScaler
```

```python
numeric_cols = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','LSTAT'] 
scaler = StandardScaler()
scaler.fit(housing_df[numeric_cols])
```

```python
housing_df
```

```python
scaled_inputs = scaler.transform(housing_df[numeric_cols])
scaled_inputs
```

```python
inputs = scaled_inputs
targets = housing_df.MEDV

# Create and train the model
model = LinearRegression().fit(inputs, targets)

# Generate predictions
predictions = model.predict(inputs)

# Compute loss to evalute the model
loss = rmse(targets, predictions)
print('Loss:', loss)
```

```python
weights_df = pd.DataFrame({
    'feature': np.append(numeric_cols, 1),
    'weight': np.append(model.coef_, model.intercept_)
})
weights_df.sort_values('weight', ascending=False)
```
